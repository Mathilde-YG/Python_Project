**5) Calculate the correlation coefficients between the Happiness_Index and the
continuous variables (economy, family, health, freedom, trust, generosity). For each
year, which variable has the highest positive correlation with happiness and which variable has the highes negative
correlation with happiness**

```python
# Define continuous variables
conti_vars = [
    'GDP per capita', # economy
    'Healthy life expectancy', # health
    'Freedom to make life choices', # freedom
    'Perceptions of corruption', # trust
    'Generosity', # genrosity
    'Family', # family
]
```

```python
# Function to calculate correlations
def calcul_corre(df):
    # Check available columns
    available_vars = [var for var in conti_vars if var in df.columns]
    if 'Score' not in df.columns:
        raise ValueError("Happiness Score column is missing in the DataFrame.")
    
    # Compute correlations
    correlations = df[available_vars + ['Score']].corr()['Score'].drop('Score')
    
    # Identify highest positive and negative correlations
    highest_pos_corr = correlations.idxmax() if not correlations.empty else None
    highest_neg_corr = correlations.idxmin() if not correlations.empty else None
    
    return correlations, highest_pos_corr, highest_neg_corr

# Calculate correlations for each year
results = {}

for year, df in {'2015': df_2015, '2016': df_2016, '2017': df_2017, '2018': df_2018, '2019': df_2019}.items():
    try:
        # Calculate correlations
        correlations, highest_pos_corr, highest_neg_corr = calcul_corre(df)
        
        results[year] = {
            'Correlations': correlations,
            'Highest Positive Correlation': highest_pos_corr,
            'Highest Negative Correlation': highest_neg_corr
        }
    except Exception as e:
        print(f"Error processing data for {year}: {e}")

# Print the results
for year, result in results.items():
    print(f"Year: {year}")
    print("Correlations with Happiness Score:")
    print(result['Correlations'])
    print(f"Highest Positive Correlation: {result['Highest Positive Correlation']}")
    print(f"Highest Negative Correlation: {result['Highest Negative Correlation']}")
    print()
```

## Modelling (Use LinearRegression model from the scikit-learn library)

**1) For each year in the dataset, create a unidimensional linear model where the
independent variable is one of the factors (economy, family, health, freedom, trust,
generosity), and the dependent variable is the happiness score. Report the coefficients
and intercept for each linear model and calculate the coefficient of determination (R^2)
for each model.**


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Define the continuous variables (factors)
conti_vars = [
    'GDP per capita', # economy
    'Healthy life expectancy', # health
    'Freedom to make life choices', # freedom
    'Perceptions of corruption', # trust
    'Generosity', # generosity
    'Family', # family
]

# Function to fit a linear model and calculate metrics
def fit_linear_model(df, factor):
    X = df[[factor]].dropna()  # Independent variable
    y = df['Score'].dropna()   # Dependent variable

    # Ensure X and y have the same length
    min_len = min(len(X), len(y))
    X = X.iloc[:min_len]
    y = y.iloc[:min_len]

    # Fit the model
    model = LinearRegression()
    model.fit(X, y)
    
    # Predict and calculate R^2
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    
    return model.coef_[0], model.intercept_, r2
```

```python
# Calculate coefficients, intercepts, and R^2 for each year and factor
results = {}

for year, df in {'2015': df_2015, '2016': df_2016, '2017': df_2017, '2018': df_2018, '2019': df_2019}.items():
    year_results = {}
    for factor in conti_vars:
        if factor in df.columns:
            coef, intercept, r2 = fit_linear_model(df, factor)
            year_results[factor] = {
                'Coefficient': coef,
                'Intercept': intercept,
                'R^2': r2
            }
    
    results[year] = year_results

# Print the results
for year, year_result in results.items():
    print(f"Year: {year}")
    for factor, metrics in year_result.items():
        print(f"Factor: {factor}")
        print(f"  Coefficient: {metrics['Coefficient']:.4f}")
        print(f"  Intercept: {metrics['Intercept']:.4f}")
        print(f"  R^2: {metrics['R^2']:.4f}")
    print()
```

```python
# Function to fit a linear model and calculate metrics
def fit_linear_model(df, factor):
    if factor not in df.columns:
        return None, None, None  # Return None if variable is not in the DataFrame
    
    X = df[[factor]].dropna()  # Independent variable
    y = df['Score'].dropna()   # Dependent variable

    # Ensure X and y have the same length
    min_len = min(len(X), len(y))
    X = X.iloc[:min_len]
    y = y.iloc[:min_len]

    # Fit the model
    model = LinearRegression()
    model.fit(X, y)
    
    # Predict and calculate R^2
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    
    return model.coef_[0], model.intercept_, r2

# Calculate coefficients, intercepts, and R^2 for each year and factor
results = {}

for year, df in {'2015': df_2015, '2016': df_2016, '2017': df_2017, '2018': df_2018, '2019': df_2019}.items():
    year_results = {}
    for factor in conti_vars:
        if factor in df.columns:
            coef, intercept, r2 = fit_linear_model(df, factor)
            if coef is not None:
                year_results[factor] = {
                    'Coefficient': coef,
                    'Intercept': intercept,
                    'R^2': r2
                }
    
    results[year] = year_results

# Print the results
for year, year_result in results.items():
    print(f"Year: {year}")
    for factor, metrics in year_result.items():
        print(f"Factor: {factor}")
        print(f"  Coefficient: {metrics['Coefficient']:.4f}")
        print(f"  Intercept: {metrics['Intercept']:.4f}")
        print(f"  R^2: {metrics['R^2']:.4f}")
    print()

# Function to find the best variable for each year
def best_variable_per_year(results):
    best_vars = {}
    for year, metrics in results.items():
        best_r2 = -1
        best_variable = None
        for factor, values in metrics.items():
            if values['R^2'] > best_r2:
                best_r2 = values['R^2']
                best_variable = factor
        best_vars[year] = {
            'Best Variable': best_variable,
            'R^2': best_r2
        }
    return best_vars

# Determine the best variable for each year
best_vars = best_variable_per_year(results)

# Print the best variables
for year, result in best_vars.items():
    print(f"Year: {year}")
    print(f"Best Variable: {result['Best Variable']}")
    print(f"R^2: {result['R^2']:.4f}")
    print()

# Check if the best variable is consistent across years
best_variables = set(result['Best Variable'] for result in best_vars.values())
if len(best_variables) == 1:
    print("The best variable is consistent across all years.")
else:
    print("The best variable varies across different years.")
```

**1) Consider the population to consist of all the countries listed. For the 2019 data,
randomly select a sample of 30 countries from the population. Compute a 95%
confidence interval for the sample mean of The Happiness score.**


```python
import numpy as np
import scipy.stats as stats

# Assume df_2019 is DataFrame containing the 2019 data.

# Randomly select 30 countries
sample = df_2019['Score'].sample(n=30, random_state=42)

# Calculate the sample mean
sample_mean = sample.mean()

# Calculate the sample standard deviation
sample_std = sample.std(ddof=1)  # ddof=1 for sample standard deviation

# Calculate the 95% confidence interval
confidence_level = 0.95
z_score = stats.norm.ppf((1 + confidence_level) / 2)  # z-score for 95% confidence

margin_of_error = z_score * (sample_std / np.sqrt(len(sample)))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Print the results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"95% Confidence Interval: {confidence_interval[0]:.4f} to {confidence_interval[1]:.4f}")
```

**2) When constructing the dataset, various variables were considered, including the
annual sunshine duration in hours for each country. Although the data was not
ultimately included, we know that the standard deviation of the values was 200 hours.
The newspaper claimed that the population average was 2500 hours or less. To
investigate this claim, a new measurement was conducted last year on a small sample of
50 countries randomly selected, and the sample average was found to be 2580 hours.
Formulate the null and alternative hypotheses to test the claim and determine whether
this should be a one-sided or two-sided test. Based on the sampled data, state whether the null hypothesis should be rejected or not**

```python
import scipy.stats as stats
import numpy as np

# Given data
sample_mean = 2580  # Sample mean (x)
population_mean = 2500  # Population mean under H0 (0)
population_std_dev = 200  # Population standard deviation ()
sample_size = 50  # Sample size (n)

# Calculate the standard error of the mean (SEM)
standard_error = population_std_dev / np.sqrt(sample_size)

# Calculate the z-score
z_score = (sample_mean - population_mean) / standard_error

# Find the critical z-value for a one-sided test at  = 0.05
alpha = 0.05
critical_z = stats.norm.ppf(1 - alpha)

# Print the results
print(f"Sample Mean: {sample_mean}")
print(f"Population Mean (H0): {population_mean}")
print(f"Population Std Dev: {population_std_dev}")
print(f"Sample Size: {sample_size}")
print(f"Standard Error: {standard_error:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"Critical Z-Value for one-sided test at  = 0.05: {critical_z:.2f}")

# Decision rule
if z_score > critical_z:
    print("Reject the null hypothesis: The population mean is greater than 2500 hours.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to conclude the population mean is greater than 2500 hours.")
```

**Problem Statement**
The newspaper's claim that the population average annual sunshine duration is 2500 hours or less. A sample of 50 countries was taken, with a sample mean of 2580 hours and a population standard deviation of 200 hours.

**Hypotheses**
Null Hypothesis : The population mean annual sunshine duration is 2500 hours or less. 
Alternative Hypothesis : The population mean annual sunshine duration is greater than 2500 hours. 

**Test type**
This is a **one-sided test** because we are specifically testing if the mean is greater than 2500 hours.

**Z-Test Calculation**
- Sample Mean: 2580
- Population Mean (H0): 2500
- Population Std Dev: 200
- Sample Size: 50

- Z-Score: 2.83
- Critical Z-Value for one-sided test at  = 0.05: 1.64

**Conclusion**
Since the calculated Z-Score (2.83) is greater than the critical Z-Value (1.64), we reject the null hypothesis. This indicates that there is enough evidence to conclude that the population mean annual sunshine duration is greater than 2500 hours.
