```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression

# Create dummy variables
onehot = OneHotEncoder(drop='first', sparse_output=False)  # Avoid multicollinearity by dropping the first category
age_encoded = onehot.fit_transform(df_1[['Age']])
age_df = pd.DataFrame(age_encoded, columns=[f'age_cat_{i}' for i in range(1, 13)])

# Apply regression model (using ordinal encoding)
X_ordinal = df_1[['Age']]  # Using the ordinal Age variable
y = df_1['BMI']

model_ordinal = LinearRegression()
model_ordinal.fit(X_ordinal, y)

# Apply regression model (using dummy variables)
X_dummy = age_df
model_dummy = LinearRegression()
model_dummy.fit(X_dummy, y)

# Interpret results
print("Results using ordinal encoding:")
print(f"Age coefficient: {model_ordinal.coef_[0]}")

print("\nResults using dummy variables:")
for i, coef in enumerate(model_dummy.coef_):
    print(f"Age category {i+2} coefficient: {coef}")  # Coefficients for each age category

# Example predictions
new_age = pd.DataFrame({'Age': [9]})  # Example: Age category 9 (e.g., 40-44 years)
new_age_ordinal = new_age  # For ordinal encoding
new_age_dummy = onehot.transform(new_age)  # For dummy variables

print("\nPrediction results:")
print(f"Prediction using ordinal encoding: {model_ordinal.predict(new_age_ordinal)[0]}")
print(f"Prediction using dummy variables: {model_dummy.predict(new_age_dummy)[0]}")
```

1. Ordinal Encoding:
The age coefficient is -0.0792, meaning that as the Age variable increases by one unit (assuming linearity), the BMI is predicted to decrease by approximately 0.0792 units. This suggests a slight negative linear relationship between age and BMI when using ordinal encoding.

2. Dummy Variables:
Each age category coefficient represents the effect of belonging to that category compared to the reference category (the first category, which was dropped).
For example, being in age category 9 (e.g., 40-44 years) increases the predicted BMI by 2.67 units compared to the reference category.
Predictions:

- Ordinal Encoding: The predicted BMI for someone in category 9 is 28.31, reflecting a direct linear relationship with the ordinal value.
- Dummy Variables: The predicted BMI is 28.75, derived by adding the coefficient for category 9 to the intercept, capturing the more nuanced differences between categories.

```python
age = df_1['Age']
bmi = df_1['BMI']

# Fitting the linear regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(age.values.reshape(-1, 1), bmi)

# Generating prediction line
age_range = np.linspace(age.min(), age.max(), 100).reshape(-1, 1)
bmi_pred = model.predict(age_range)

# Drawing the graph
plt.figure(figsize=(10, 6))
plt.scatter(age, bmi, alpha=0.5)
plt.plot(age_range, bmi_pred, color='red', linewidth=2)
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('BMI vs Age Linear Regression')
plt.grid(True)
plt.show()

# Printing results
print(f"Slope: {model.coef_[0]:.4f}")
print(f"Intercept: {model.intercept_:.4f}")
print(f"R-squared: {model.score(age.values.reshape(-1, 1), bmi):.4f}")
```

- Slope: -0.0792 indicates that for every 1 unit increase in age, BMI decreases by approximately 0.0792 units. However, this value suggests that the relationship between age and BMI is weak.

- Intercept: 29.0188 represents the expected BMI value when age is 0. However, since an age of 0 has no practical meaning in the data, this value is difficult to interpret.

- R-squared: 0.0013 means that the model explains very little of the variability in the data. In other words, it indicates that the relationship between age and BMI is very weak. An R-squared value close to 0 signifies that the model does not explain the data well.

- Graph Interpretation: As seen in the graph, BMI values are densely clustered within a specific range (20-100), and there is no clear trend based on age. This suggests that age does not significantly affect BMI.

In conclusion, the linear regression analysis between age and BMI is inappropriate. Due to the weak relationship between age and BMI, it would be better to consider other variables or a nonlinear model.

# Logistic Regression Model

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression


# Preprocessing for df_2
df_2_encoded = pd.get_dummies(df_2, drop_first=True)  # One-hot encoding for categorical variables
X_df_2 = df_2_encoded.drop('Diabetes_binary', axis=1)  # Features (independent variables)
y_df_2 = df_2_encoded['Diabetes_binary']  # Target variable (dependent variable)

# Scaling
scaler = StandardScaler()  # Initialize the StandardScaler
X_df_2_scaled = scaler.fit_transform(X_df_2)  # Fit and transform the features

# Model training
model = LogisticRegression()  # Initialize the Logistic Regression model
model.fit(X_df_2_scaled, y_df_2)  # Fit the model to the scaled features and target variable

# Preprocessing for df_3
df_3_encoded = pd.get_dummies(df_3, drop_first=True)  # One-hot encoding for categorical variables

# Select feature columns used in df_2
feature_columns = X_df_2.columns  # Get the feature columns from df_2

# Select the same features from df_3, filling missing features with 0
X_df_3 = df_3_encoded.reindex(columns=feature_columns, fill_value=0)  # Align columns and fill missing with 0

# Scaling
X_df_3_scaled = scaler.transform(X_df_3)  # Scale the features from df_3 using the previously fitted scaler

# Prediction
y_pred_df_3 = model.predict(X_df_3_scaled)  # Use the model to make predictions on the scaled df_3 features

# Check prediction results
df_3['Predicted_Diabetes'] = y_pred_df_3  # Add the predictions to df_3
print(df_3[['Age', 'Sex', 'Predicted_Diabetes']])  # Print the relevant columns including predictions
```

```python
from sklearn.metrics import confusion_matrix
# Assuming df_3 has the actual target variable 'Diabetes_binary'
y_df_3 = df_3['Diabetes_binary']  # Define the actual target variable for df_3

# Create the Confusion Matrix
conf_matrix = confusion_matrix(y_df_3, y_pred_df_3)

# Visualize the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Non-diabetes', 'Pre-diabetes/Diabetes'], 
            yticklabels=['Non-diabetes', 'Pre-diabetes/Diabetes'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Class')
plt.ylabel('True Class')
plt.show()

```

- Values in Each Cell:

    - 158,201: The number of actual Non-diabetes cases that the model correctly predicted as Non-diabetes (True Negative).
    - 60,133: The number of actual Non-diabetes cases that the model incorrectly predicted as Pre-diabetes/Diabetes (False Positive).
    - 8,210: The number of actual Pre-diabetes/Diabetes cases that the model incorrectly predicted as Non-diabetes (False Negative).
    - 27,136: The number of actual Pre-diabetes/Diabetes cases that the model correctly predicted as Pre-diabetes/Diabetes (True Positive).

```python
# Assuming 'BMI' is one of the features in your model
# Create a range of BMI values for plotting
bmi_range = np.linspace(X_df_2['BMI'].min(), X_df_2['BMI'].max(), 500).reshape(-1, 1)

# Recreate a DataFrame to include the necessary features
# Assuming 'Age' and 'Sex' are also features, you'll need to fix them accordingly
# Here, we will create a dummy DataFrame for demonstration
# Replace 'mean_age' and 'mode_sex' with appropriate values from your dataset
mean_age = df_2['Age'].mean()
mode_sex = df_2['Sex'].mode()[0]  # Most common value

# Create a DataFrame for predictions
bmi_df = pd.DataFrame({
    'BMI': bmi_range.flatten(),
    'Age': mean_age,
    'Sex': mode_sex
})

# One-hot encode the new DataFrame
bmi_encoded = pd.get_dummies(bmi_df, drop_first=True)

# Align columns with the training data
bmi_encoded = bmi_encoded.reindex(columns=feature_columns, fill_value=0)

# Scale the features
bmi_scaled = scaler.transform(bmi_encoded)

# Predict probabilities
probs = model.predict_proba(bmi_scaled)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(bmi_range, probs[:, 0], label='Class 0: No Diabetes', color='blue')
plt.plot(bmi_range, probs[:, 1], label='Class 1: Diabetes', color='red')
plt.title(f'Logistic Regression: Predicted Probabilities for Diabetes based on BMI (Age Category: {mean_age}, Sex: {mode_sex})')
plt.xlabel('BMI')
plt.ylabel('Probability')
plt.legend()
plt.grid()
plt.show()
```

1. Axes Description:
- X-axis (BMI): This represents the Body Mass Index (BMI) values. Generally, higher BMI values are linked to increased health risks related to obesity.
- Y-axis (Probability): This indicates the predicted probabilities of having diabetes (Class 1) or not having diabetes (Class 0), ranging from 0 to 1.

2. Line Colors:
- Blue Line (Class 0: No Diabetes): The probability of not having diabetes is represented by the blue line (Class 0: No Diabetes). ## ANOVA

```python
import scipy.stats as stats

# ANOVA analysis: Relationship between Mental Health Days (MentHlth) and Diabetes_012
# Assuming df_1 contains MentHlth and Diabetes_012
groups = [df_1[df_1['Diabetes_012'] == i]['MentHlth'] for i in range(3)]

# Perform ANOVA analysis
f_stat, p_value = stats.f_oneway(*groups)

# Print results
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

# Interpret P-value
if p_value < 0.05:
    print("There is a significant difference (reject the null hypothesis).")
else:
    print("There is no difference (fail to reject the null hypothesis).")

# We can conclude that there is a significant difference in Mental Health Days (MentHlth) among the diabetes status groups.
```

1. F-statistic: 717.12:
- The F-statistic of approximately 717.12 indicates a relatively high ratio of variance between the groups compared to the variance within the groups.
- This high value indicates that there is a substantial difference between the means of the different diabetes status groups when it comes to Mental Health Days (MentHlth).

2. P-value: 2.74e-311:
- The P-value of approximately 2.74 × 10^-311 is extremely small, effectively close to zero.
- According to this P-value, there is essentially no chance of seeing an F-statistic this large (or larger) under the null hypothesis, which holds that there are no differences between the group averages.

3. Conclusion:
- Significant Difference: Since the P-value is much less than the conventional threshold of 0.05, there is a strong evidence to reject the null hypothesis.
- Implication: It concludes that there is a statistically significant difference in Mental Health Days among the different diabetes status groups (0, 1, and 2). This means that individuals in these groups experience different numbers of mental health days, suggesting that diabetes status has a significant effect on mental health.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Ensure 'Diabetes_012' is treated as a categorical variable
df_1['Diabetes_012'] = df_1['Diabetes_012'].astype('category')

# Define and fit the model: MentHlth as DV, Diabetes_012 as IV
model = ols('MentHlth ~ C(Diabetes_012)', data=df_1).fit()

# Generate ANOVA table
anova_table = sm.stats.anova_lm(model, typ=1)

# Print ANOVA table
print(anova_table)

# Interpret P-value from the ANOVA table
p_value = anova_table['PR(>F)'][0]
if p_value < 0.05:
    print("There is a significant difference (reject the null hypothesis).")
else:
    print("There is no significant difference (fail to reject the null hypothesis).")
```

- Here, the one-way ANOVA is used for analysing the difference between the groups of 'Diabetes_012'.
- Null Hypothesis (H₀): There is no significant difference in the mean values of 'MentHlth' across the 'Diabetes_012' groups (0, 1, 2).
- Alternative Hypothesis (H₁): There is a significant difference in the mean values of MentHlth among the Diabetes_012 groups.
- The p-value is under 0.05 (5%).
- The result shows that there is a significant difference in the mean values of 'MenHlth' among the Diabetes_012 groups.

```python
# One-hot encoding
df_2_encoded = pd.get_dummies(df_2, drop_first=True)  # Convert categorical variables into dummy/indicator variables, dropping the first category to avoid multicollinearity.

# Scaling continuous variables (Normalization)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()  # Initialize the MinMaxScaler
# Fit the scaler to the features and transform them
scaled_features = scaler.fit_transform(df_2_encoded.drop('Diabetes_binary', axis=1))  
# Create a new DataFrame with the scaled features
df_2_scaled = pd.DataFrame(scaled_features, columns=df_2_encoded.columns[:-1])  
# Add the target variable back to the scaled DataFrame
df_2_scaled['Diabetes_binary'] = df_2_encoded['Diabetes_binary']  
```

```python
from sklearn.model_selection import train_test_split  # Import the train_test_split function
from sklearn.ensemble import RandomForestClassifier  # Import the RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix  # Import metrics for evaluation

# Split the data into training and testing sets
X = df_2_scaled.drop('Diabetes_binary', axis=1)  # Features (independent variables)
y = df_2_scaled['Diabetes_binary']  # Target variable (dependent variable)

# Perform the train-test split, using 30% of the data for testing and stratifying by the target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Train the model using Random Forest
model = RandomForestClassifier(random_state=42)  # Initialize the Random Forest classifier
model.fit(X_train, y_train)  # Fit the model to the training data
```

- RandomForestClassifier:
This is a machine learning model that belongs to the ensemble learning family. It constructs multiple decision trees during training and outputs the mode of the classes (for classification) or the mean prediction (for regression) of the individual trees.
Random forests are often used because they are robust against overfitting and can handle a mix of numerical and categorical features.

- random_state=42:
The random_state parameter is used to control the randomness of the model. By setting it to a specific integer (in this case, 42), you ensure that the results are reproducible. This means that every time you run the code with this random_state, you will get the same results.
Using a fixed random_state is particularly useful for debugging and comparing different models, as it helps to eliminate variability caused by random splits of the data or random selections within the algorithm.

```python
# Prediction
y_pred = model.predict(X_test)  # Use the trained model to make predictions on the test set

# Performance evaluation
accuracy = accuracy_score(y_test, y_pred)  # Calculate the accuracy of the model
print(f"Model Accuracy: {accuracy:.2f}")  # Print the accuracy formatted to two decimal places

# Classification report
print(classification_report(y_test, y_pred))  # Print a detailed classification report

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)  # Generate the confusion matrix
plt.figure(figsize=(8, 6))  # Set the figure size
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Non-diabetes', 'Pre-diabetes/Diabetes'], 
            yticklabels=['Non-diabetes', 'Pre-diabetes/Diabetes'])  # Create a heatmap for the confusion matrix
plt.title('Confusion Matrix')  # Set the title of the plot
plt.xlabel('Prediction Class')  # Set the x-axis label
plt.ylabel('Real Class')  # Set the y-axis label
plt.show()  # Display the plot
```
