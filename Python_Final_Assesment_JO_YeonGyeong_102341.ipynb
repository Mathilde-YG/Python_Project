```python
import statsmodels.api as sm

# Define independent variables (multiple predictors)
X = df_math[['freetime', 'goout', 'health', 'failures']]  # Multiple independent variables
y = df_math['G3']

# Add a constant term for the intercept
X = sm.add_constant(X)

# Fit the OLS model
model = sm.OLS(y, X).fit()  # Fitting the model
print(model.summary())  # Output the summary of the regression results
```

### Interpretation for statistic results of math data
#### Independent Variable (IV) and Dependent Variable (DV)
- Dependent Variable (DV): The variable being predicted, which is G3 (the final grades of students).
- Independent Variables (IV): The factors that may influence the dependent variable in this analysis:
**freetime**: The amount of free time students have.
**goout**: The frequency of going out.
**health**: The health status of students.
**failures**: The number of class failures.
  
#### R-squared
- R-squared: The R-squared value is 0.145, meaning that approximately 14.5% of the variance in G3 can be explained by the independent variables included in the model.

#### Coefficients (coef) and Intercept (const)
- Intercept (const): The intercept value is 11.9298, representing the expected value of G3 when all independent variables are equal to zero. This serves as a baseline score for the model.
- Coefficients:
- **freetime (0.3620)**: For each unit increase in free time, G3 is expected to increase by approximately 0.3620 points, assuming all other variables remain constant. However, the effect is not statistically significant (p-value = 0.109).
- **goout (-0.4611)**: For each unit increase in the frequency of going out, G3 is expected to decrease by approximately 0.4611 points, which is statistically significant (p-value = 0.023).
- **health (-0.1493)**: For each unit increase in health status, G3 is expected to decrease by approximately 0.1493 points, but this effect is not statistically significant (p-value = 0.337).
- **failures (-2.1608)**: For each additional failure, G3 is expected to decrease by approximately 2.1608 points, which is statistically significant (p-value = 0.000).

  
#### P-values
- P-values: These values help determine the statistical significance of each coefficient, with a common threshold for significance set at 0.05.

- **freetime (p-value = 0.305)**: Not statistically significant (greater than 0.05), suggesting that free time does not have a significant effect on G3.
- **goout (p-value = 0.023)**: Statistically significant (less than 0.05), indicating that going out has a negative effect on G3.
- **health (p-value = 0.337)**: Not statistically significant, suggesting that health status does not significantly affect G3.
- **failures  (p-value = 0.000)**: Statistically significant, indicating a significant negative effect on G3, meaning that an increase in failures is linked to lower grades.

### Probit Regression with 'success' as the Dependent Variable and 'freetime,' 'health,' and 'failures' as Independent Variables

```python
# Create the dependent variable for Probit regression (1: success, 0: failure)
df_math['success'] = np.where(df_math['G3'] >= 10, 1, 0)  # Success : if G3 is 10 or higher
df_por['success'] = np.where(df_por['G3'] >= 10, 1, 0)    # Success : if G3 is 10 or higher

# Set independent variables
X_math = df_math[['freetime', 'health', 'failures']]
y_math = df_math['success']

X_por = df_por[['freetime', 'health', 'failures']]
y_por = df_por['success']

# Add a constant term
X_math = sm.add_constant(X_math)
X_por = sm.add_constant(X_por)

# Fit the Probit model for the math subject
model_math = sm.Probit(y_math, X_math).fit()
print("Math Probit Regression Summary:")
print(model_math.summary())

# Fit the Probit model for the Portuguese subject
model_por = sm.Probit(y_por, X_por).fit()
print("\nPortuguese Probit Regression Summary:")
print(model_por.summary())
```

### Probit Regression Model Summary

#### Model Performance
- **Math Model**: About 8.79% of the variability is explained by the pseudo R-squared of 0.08787. -228.26 is the log-likelihood.

- DV : 'success' ('success' is binomial, so it will be adept to use the Logistic Regression Model)

```python
# re-read the data
df_math = pd.read_csv('student-mat.csv', sep=';', encoding='cp1252')
df_math.head()
df_por = pd.read_csv('student-por.csv', sep=';', encoding='cp1252')
df_por.head()
```

```python
# For Math dataset
# Create 'success' column: 1 if G3 >= 10, otherwise 0
df_math['success'] = (df_math['G3'] >= 10).astype(int)

# Convert gender variable to binary (0 for female, 1 for male)
df_math['sex'] = df_math['sex'].replace({'F': 0, 'M': 1}).astype(int)

# Set independent variables
X_math = df_math[['failures', 'studytime', 'absences', 'sex']]
y_math = df_math['success']

# Add constant term
X_math = sm.add_constant(X_math)

# Fit logistic regression model
math_model = sm.Logit(y_math, X_math).fit()

# Print results for Math model
print("Math Model Summary")
print(math_model.summary())

# For Portuguese dataset
# Create 'success' column: 1 if G3 >= 10, otherwise 0
df_por['success'] = (df_por['G3'] >= 10).astype(int)

# Convert gender variable to binary (0 for female, 1 for male)
df_por['sex'] = df_por['sex'].replace({'F': 0, 'M': 1}).astype(int)

# Set independent variables
X_por = df_por[['failures', 'studytime', 'absences', 'sex']]
y_por = df_por['success']

# Add constant term
X_por = sm.add_constant(X_por)

# Fit logistic regression model
por_model = sm.Logit(y_por, X_por).fit()

# Print results for Portuguese model
print("\nPortuguese Model Summary")
print(por_model.summary())
```
